## üß© What are Functions?
A **function** is a reusable block of code that performs a specific task.  
Functions help make programs **modular, organized, and easier to debug**.

````python
def greet(name):
    return f"Hello, {name}!"
    
print(greet("Umer"))
````

## Variable Scope

Variable scope determines where a variable can be accessed in the program.

There are two main types:

1- **Local Variables** ‚Äì declared inside a function; accessible only within that function.

2- **Global Variables** ‚Äì declared outside all functions; accessible anywhere in the code.

````python
x = 10  # Global variable

def show_scope():
    x = 5  # Local variable
    print("Inside function, x =", x)

show_scope()
print("Outside function, x =", x)
````

````python
count = 0  # Global variable

def increase_count():
    global count
    count += 1
    print("Inside function, count =", count)

increase_count()
print("Outside function, count =", count)
````

## Lambda Expressions, Map, and Filter

## ‚ö° Lambda Expressions (Anonymous Functions)

A **lambda function** is a small, anonymous function defined using the `lambda` keyword.  
It can take **any number of arguments**, but can only have **one expression**.

````python
square = lambda x: x**2
print(square(5))
````

##  üí° When to Use Lambda Functions

- When you need a simple, one-line function

- Commonly used with **map()**, **filter()**, and **reduce()**

- Helps keep code clean and concise

##  üß© The **map()** Function

The **map()** function applies a given function to each item in an iterable (like a list).

```python
map(function, iterable)



````python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
````

##  üîç The **filter()** Function

The **filter()** function filters elements from an iterable based on a condition (True/False).

### Syntax
``` python 
filter(function, iterable)

````python
numbers = [10, 15, 20, 25, 30]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
````

##  ‚öôÔ∏è Combining **map()** and **filter()**

You can combine them for more complex tasks.

````python
numbers = [1, 2, 3, 4, 5, 6]
# Square only even numbers
result = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)))
print(result)
````

## üß© What is an Inner (Nested) Function?

An **inner function** (also called a **nested function**) is a function **defined inside another function**.  
They help **organize code**, create **encapsulation**, and enable the use of **closures**.



## ‚úÖ Syntax

```python
def outer_function():
    def inner_function():
        # code block
        pass
    inner_function()  # calling inner function

````python
def outer():
    print("Outer function starts")

    def inner():
        print("Inner function runs")

    inner()  # calling the inner function

    print("Outer function ends")

outer()
````

##  üéØ Why Use Inner Functions?

- To logically group functionality that‚Äôs only used inside another function

- To hide implementation details (encapsulation)

- To create closures that remember state

##  üìò Example 2: Returning an Inner Function

You can return the inner function itself instead of calling it.

This creates a closure ‚Äî the inner function remembers the variables of the outer function.

````python
def outer(msg):
    def inner():
        print("Message:", msg)
    return inner

my_func = outer("Hello from inner function!")
my_func()
````

- finally always runs (for cleanup like closing files).

## üß© What are Classes and Objects?
- A **Class** is a blueprint for creating objects.  
- An **Object** is an instance of a class that holds data (variables) and behavior (methods).



## ‚öôÔ∏è Defining a Class
A class contains:
- **Instance Variables** ‚Üí Data unique to each object.  
- **Methods** ‚Üí Functions defined inside a class that operate on its data.

### ‚úÖ Syntax
```python
class ClassName:
    def __init__(self, parameters):
        self.variable = value  # instance variable

    def method_name(self):
        # method body

##  üß† Understanding __init__ Method

-   The __init__() method is the constructor.

-   It runs automatically when an object is created.

-   It initializes the object‚Äôs variables.

````python
class Car:
    def __init__(self, brand, model, year):
        # Instance variables
        self.brand = brand
        self.model = model
        self.year = year

    # Instance method
    def display_info(self):
        print(f"Car: {self.brand} {self.model} ({self.year})")

    # Another instance method
    def start_engine(self):
        print(f"The {self.brand} {self.model}'s engine has started.")

# Creating objects (instances)
car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Honda", "Civic", 2022)

# Accessing instance methods
car1.display_info()
car2.display_info()

car1.start_engine()
car2.start_engine()
````

##  Class Variables and Functions, Constructors and Destructors**

## üß© What are Class Variables?

- **Class Variables** are shared by **all objects** of a class.  
- Unlike instance variables (unique to each object), class variables belong to the **class itself**.  
- They are defined **outside all methods** but **inside the class definition**.

### ‚úÖ Example Syntax
```python
class ClassName:
    class_variable = value
    def __init__(self, instance_var):
        self.instance_var = instance_var

##  ‚öôÔ∏è Class Functions (Class Methods)

-   A class method is defined using the @classmethod decorator.

-   It takes cls (the class itself) as its first parameter instead of self.

-   It can access and modify class variables but not instance variables.

````python
@classmethod
def method_name(cls):
    # method body
    pass
````

##  üß± Constructors (__init__)

-   The constructor in Python is the __init__() method.

-   It is called automatically when an object is created.

-   It initializes instance variables for that object.

##  üßπ Destructors (__del__)

-   The destructor in Python is the __del__() method.

-   It is called automatically when an object is destroyed (usually at the end of program execution or when explicitly deleted).

-   Used to perform cleanup operations (like closing files or releasing resources).

````python
class Student:
    # Class variable (shared by all instances)
    school_name = "ABC High School"

    # Constructor
    def __init__(self, name, grade):
        self.name = name        # Instance variable
        self.grade = grade      # Instance variable
        print(f"Student {self.name} has been created.")

    # Instance method
    def show_details(self):
        print(f"Name: {self.name}, Grade: {self.grade}, School: {Student.school_name}")

    # Class method
    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name
        print(f"School name changed to: {cls.school_name}")

    # Destructor
    def __del__(self):
        print(f"Student {self.name} record deleted.")

# Creating objects
s1 = Student("Umer", "A")
s2 = Student("Ali", "B")

# Display details
s1.show_details()
s2.show_details()

# Modify class variable using class method
Student.change_school("Global Grammar School")

# Display updated details
s1.show_details()
s2.show_details()

# Delete object to trigger destructor
del s1
````

##  üß≠ Summary:

| Concept               | Description                 | Keyword               |
| --------------------- | --------------------------- | --------------------- |
| **Class Variable**    | Shared among all objects    | Defined inside class  |
| **Instance Variable** | Unique to each object       | Defined in `__init__` |
| **Class Method**      | Operates on class variables | `@classmethod`        |
| **Constructor**       | Initializes object          | `__init__()`          |
| **Destructor**        | Destroys object, cleanup    | `__del__()`           |


##  Inheritance and Multilevel Inheritance

## üß© What is Inheritance?

**Inheritance** is an Object-Oriented Programming (OOP) concept that allows a **child class** to acquire the **properties and behaviors (methods)** of a **parent class**.

It promotes **code reusability**, **scalability**, and **organization**.



## üß± Types of Inheritance in Python
1. **Single Inheritance** ‚Äì One base class and one derived class  
2. **Multiple Inheritance** ‚Äì Derived class inherits from more than one base class  
3. **Multilevel Inheritance** ‚Äì A derived class acts as a base for another class  
4. **Hierarchical Inheritance** ‚Äì Multiple derived classes inherit from one base class  
5. **Hybrid Inheritance** ‚Äì Combination of multiple inheritance types  



## ‚öôÔ∏è Multilevel Inheritance

In **multilevel inheritance**, a class is derived from another **derived class**.  
This forms a **chain of inheritance**.



### 1 ‚Äì Single Inheritance

````python
# Single Inheritance Example

class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

dog = Dog()
dog.sound()   # Inherited from Animal
dog.bark()    # Defined in Dog
````

### 2 ‚Äì Multiple Inheritance

````python
# Multiple Inheritance Example

class Father:
    def skill(self):
        print("Father: Driving")

class Mother:
    def hobby(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    def talent(self):
        print("Child: Coding")

c = Child()
c.skill()     # From Father
c.hobby()     # From Mother
c.talent()    # Own method
````

### 3 ‚Äì Multilevel Inheritance

````python
# Multilevel Inheritance Example

class Grandparent:
    def legacy(self):
        print("Grandparent: Land owner")

class Parent(Grandparent):
    def property(self):
        print("Parent: Car owner")

class Child(Parent):
    def dream(self):
        print("Child: Wants to build empire")

child = Child()
child.legacy()    # From Grandparent
child.property()  # From Parent
child.dream()     # Own method
````

### 4 ‚Äì Hierarchical Inheritance

````python
# Hierarchical Inheritance Example

class Vehicle:
    def engine(self):
        print("All vehicles have engines")

class Car(Vehicle):
    def four_wheels(self):
        print("Car has 4 wheels")

class Bike(Vehicle):
    def two_wheels(self):
        print("Bike has 2 wheels")

c = Car()
b = Bike()

c.engine()
c.four_wheels()

b.engine()
b.two_wheels()
````

### 5 ‚Äì Hybrid Inheritance

````python
# Hybrid Inheritance Example

class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(A):
    def showC(self):
        print("Class C")

class D(B, C):
    def showD(self):
        print("Class D")

obj = D()
obj.showA()   # From A
obj.showB()   # From B
obj.showC()   # From C
obj.showD()   # From D
````

## üßæ Summary: Inheritance Types in Python

| **Inheritance Type** | **Structure / Relationship** | **Example Classes** | **Key Point** |
|-----------------------|------------------------------|----------------------|----------------|
| **Single Inheritance** | One parent ‚Üí One child | `Animal ‚Üí Dog` | Simplest form; child inherits from one parent |
| **Multiple Inheritance** | Multiple parents ‚Üí One child | `Father + Mother ‚Üí Child` | Combines features from multiple parents |
| **Multilevel Inheritance** | Grandparent ‚Üí Parent ‚Üí Child | `Grandparent ‚Üí Parent ‚Üí Child` | Inheritance chain through levels |
| **Hierarchical Inheritance** | One parent ‚Üí Multiple children | `Vehicle ‚Üí Car, Bike` | Multiple subclasses share same parent |
| **Hybrid Inheritance** | Combination of multiple types | `A ‚Üí B, C ‚Üí D(B, C)` | Complex mix, often resolved by MRO (Method Resolution Order) |

##  Method Resolution Order (MRO)

## ‚öôÔ∏è What is Method Resolution Order?
- **MRO (Method Resolution Order)** defines the order in which Python looks for a method or attribute in a hierarchy of classes.
- It determines **which method gets called first** when a class is derived from multiple parent classes.
- Python uses the **C3 Linearization algorithm** to compute the MRO.



## üß© Why is MRO Important?
- In multiple inheritance, the same method might exist in more than one parent class.  
- MRO ensures a **consistent and predictable** method lookup order.  
- It avoids ambiguity and maintains a clear inheritance chain.



## üß† How to View the MRO
You can check the MRO of any class using:
1. `ClassName.__mro__`
2. `ClassName.mro()`



## üß≠ Example MRO Rule
If a class `D` inherits from classes `B` and `C`, which both inherit from `A`,  
then the lookup order for methods is:

````python
# Example: Method Resolution Order (MRO)

class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")

class C(A):
    def show(self):
        print("Class C")

class D(B, C):
    pass

# Create object of D
obj = D()

# Call method
obj.show()

# Display Method Resolution Order
print("\nMethod Resolution Order (MRO):")
print(D.__mro__)
````

##  Access Specifiers (Public, Protected, Private & Name Mangling)

## üîê What are Access Specifiers?
Access Specifiers in Python define the **visibility** or **accessibility** of class members (variables and methods).  
They help in **encapsulation**, which means hiding internal details and exposing only what‚Äôs necessary.

Python doesn‚Äôt have strict access control like Java or C++,  
but it uses **naming conventions** to indicate the intended level of access.



## üß© Types of Access Specifiers

| **Access Type** | **Syntax** | **Accessible From** | **Description** |
|------------------|-------------|----------------------|------------------|
| **Public** | `var_name` | Anywhere (inside or outside the class) | Default access level |
| **Protected** | `_var_name` | Within the class and its subclasses | Indicates internal use, but not enforced |
| **Private** | `__var_name` | Only within the class | Not directly accessible from outside (name mangling applies) |



## üß† Name Mangling
- Python internally changes the name of private attributes to prevent direct access.
- For an attribute named `__name`, Python changes it to `_ClassName__name`.
- This mechanism is called **Name Mangling**.

‚úÖ Example:  
If a class `Student` has a private variable `__marks`,  
it can still be accessed as `_Student__marks` (though not recommended).



## üí° Why Use Access Specifiers?
- To **control data access** and maintain data integrity.  
- To **hide internal implementation** details.  
- To **promote encapsulation** in object-oriented design.



## üß≠ Summary
| **Specifier** | **Example** | **Access Level** | **Purpose** |
|----------------|--------------|------------------|--------------|
| **Public** | `self.name` | Fully accessible | General use |
| **Protected** | `self._age` | Semi-restricted | For subclasses |
| **Private** | `self.__salary` | Highly restricted | Internal use only |
| **Name Mangling** | `_ClassName__variable` | Internal alias | Avoid accidental access |



````python
# Example: Public, Protected, Private Variables and Name Mangling

class Employee:
    def __init__(self, name, age, salary):
        self.name = name             # Public variable
        self._age = age              # Protected variable (by convention)
        self.__salary = salary       # Private variable

    def display_info(self):
        print(f"Name: {self.name}, Age: {self._age}")

    def __show_salary(self):         # Private method
        print(f"Salary: {self.__salary}")

# Create object
emp = Employee("Alice", 30, 70000)

# Accessing public variable
print("Public:", emp.name)

# Accessing protected variable (possible, but not recommended)
print("Protected:", emp._age)

# Trying to access private variable directly (will cause error)
# print(emp.__salary)   # ‚ùå AttributeError

# Accessing private variable using name mangling
print("Private via name mangling:", emp._Employee__salary)

# Accessing private method using name mangling
emp._Employee__show_salary()
````

##  Inner / Nested Classes (Association, Aggregation, Composition)

## üß© What is an Inner or Nested Class?
- An **Inner (Nested) Class** is a class defined **inside another class**.  
- It is used to **logically group** classes that are only used in one place.  
- It improves **code organization** and **encapsulation**.

Example:
```python
class Outer:
    class Inner:
        pass

##  üß† Why Use Inner Classes?

-   To represent a "part-of" relationship (Composition).

-   To logically group helper classes within a main class.

-   To maintain encapsulation when one class is used only by another.

##  üß≠ Relationship Types Between Classes

### 1Ô∏è‚É£ Association

-   It represents a loose relationship between two classes.

-   Both classes can exist independently of each other.

-   Example: Teacher ‚Üî Department (a teacher can exist even without a department).

### 2Ô∏è‚É£ Aggregation

-   It represents a "has-a" relationship.

-   One class contains another class‚Äôs object, but both can still exist separately.

-   Example: Car has an Engine (the engine can exist outside the car).

### 3Ô∏è‚É£ Composition

-   It represents a strong "part-of" relationship.

-   One class owns the other.

-   If the container object is destroyed, the contained object is also destroyed.

-   Example: Human has a Heart (the heart cannot exist without the human).

##  ‚öôÔ∏è Inner Class Example Use-Cases

- A University class containing a nested Department class.

- A Computer class containing a nested Processor class.

##  üßæ Summary:

| **Type**        | **Description**                    | **Object Lifespan** | **Example Relationship** |
| --------------- | ---------------------------------- | ------------------- | ------------------------ |
| **Association** | Loose connection between classes   | Independent         | Teacher ‚Üî Department     |
| **Aggregation** | One class owns another temporarily | Separate            | Car ‚Üí Engine             |
| **Composition** | Strong ownership ("part-of")       | Dependent           | Human ‚Üí Heart            |




‚úÖ Tip: Use nested classes for strong, structural relationships that make sense only inside the outer class.

#### **Example 1: Association**

````python
class Department:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, name, department):
        self.name = name
        self.department = department  # Associated, not owned

    def show(self):
        print(f"Teacher: {self.name}, Department: {self.department.name}")

# Both can exist independently
d1 = Department("Computer Science")
t1 = Teacher("Alice", d1)
t1.show()
````

### **Example 2: Aggregation**

````python
# Aggregation Example: Car and Engine

class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Car:
    def __init__(self, brand, engine):
        self.brand = brand
        self.engine = engine  # Aggregation (Car "has an" Engine)

    def show_details(self):
        print(f"Car: {self.brand}, Engine: {self.engine.horsepower} HP")

engine = Engine(150)
car = Car("Toyota", engine)
car.show_details()
````

### **Example 3: Composition**

````python
# Composition Example: Human and Heart

class Heart:
    def __init__(self):
        print("Heart created")

class Human:
    def __init__(self, name):
        self.name = name
        self.heart = Heart()  # Composition (Heart cannot exist without Human)
        print(f"Human {self.name} created")

# When Human is deleted, Heart is also deleted
h1 = Human("John")
````

### **Example 4: Nested Class**

````python
# Inner (Nested) Class Example

class Computer:
    def __init__(self, brand):
        self.brand = brand
        self.processor = self.Processor()  # Creating object of inner class

    def show_info(self):
        print(f"Computer Brand: {self.brand}")
        self.processor.show_specs()

    class Processor:   # Nested Class
        def __init__(self):
            self.cores = 8
            self.brand = "Intel"

        def show_specs(self):
            print(f"Processor: {self.brand}, Cores: {self.cores}")

# Object of outer class
comp = Computer("Dell")
comp.show_info()
````

##  Polymorphism and Operator Overloading

## üß© What is Polymorphism?
- **Polymorphism** means **"many forms"**.  
- In Python, it allows different classes to have methods with the **same name** but different behaviors.  
- The method that gets called depends on the **type of object** calling it.

Example:
```python
class Cat:
    def sound(self):
        print("Meow")

class Dog:
    def sound(self):
        print("Bark")

##  ‚öôÔ∏è Operator Overloading

### Common Dunder Methods:

### **Example 1: Polymorphism using Common Interface**

````python
class Cat:
    def sound(self):
        print("Meow")

class Dog:
    def sound(self):
        print("Bark")

class Cow:
    def sound(self):
        print("Moo")

# Polymorphism in action
animals = [Cat(), Dog(), Cow()]
for animal in animals:
    animal.sound()
````

### **Example 2: Method Overriding**

````python
class Parent:
    def greet(self):
        print("Hello from Parent class")

class Child(Parent):
    def greet(self):  # Overriding the parent method
        print("Hello from Child class")

obj1 = Parent()
obj2 = Child()

obj1.greet()  # Parent version
obj2.greet()  # Overridden version
````

### **Example 3: Operator Overloading**

````python
# Example: Adding two objects using __add__()

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Create objects
p1 = Point(2, 3)
p2 = Point(4, 5)

# Add using overloaded '+'
result = p1 + p2
print(result)
````

### **Example 4: Comparison Operator Overloading**

````python
# Example: Overloading '>' and '=='

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __gt__(self, other):  # Greater than
        return self.marks > other.marks

    def __eq__(self, other):  # Equality
        return self.marks == other.marks

s1 = Student("Alice", 85)
s2 = Student("Bob", 90)

print(s1 > s2)   # False
print(s1 == s2)  # False
````

## üß† Purpose of Magic Methods

## üß© Most Common Magic Methods Explained

- Constructor, called when an object is created.

### 5Ô∏è‚É£ `__len__(self)`

### 6Ô∏è‚É£ `__call__(self)`

### 7Ô∏è‚É£ `__del__(self)`

## üí° Why Use Dunder Methods?
‚úÖ They make custom objects behave **like built-in types**.

### **Example 1: Basic Magic Methods**

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Point({self.x}, {self.y})"

p1 = Point(3, 5)
p2 = Point(2, 4)
print(p1 + p2)  # Calls __add__()
````

## Dynamic Polymorphism (Subclass as Base Class)

## üß© What is Dynamic Polymorphism?
- **Dynamic Polymorphism** means that the **method call is resolved at runtime**, not at compile time.  
- In Python, it‚Äôs usually implemented through **method overriding** ‚Äî where a **child class redefines** a method of its **parent class**.  
- The correct version of the method (parent or child) is chosen **dynamically at runtime**, depending on the object type.



## üß† Key Concept
- A **base class reference (or variable)** can refer to an **object of a subclass**.  
- When a method is called through this reference, the **subclass version** of the method executes ‚Äî not the base class one.  

This is why it‚Äôs called **‚ÄúDynamic‚Äù** or **‚ÄúRuntime‚Äù** polymorphism.



## ‚öôÔ∏è How It Works
1. Define a **base class** with a method.
2. Create a **subclass** that overrides that method.
3. Use a **base class reference** to call the method.
4. At runtime, Python determines **which method to invoke** based on the actual object type.



## üí° Example Use Cases
- GUI frameworks where the same function (e.g., `draw()`) behaves differently for `Circle`, `Rectangle`, or `Line`.
- Game development (e.g., `move()` method for different character types).
- Data processing (e.g., `process()` method for multiple file types).






### **Example 1: Subclass as Base Class Reference**


````python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Base class reference pointing to subclass objects
animals = [Dog(), Cat(), Animal()]

for a in animals:
    a.speak()  # Method call resolved at runtime (Dynamic Polymorphism)

````

### **Example 2: Using Function That Accepts Base Class**

````python
class Vehicle:
    def start(self):
        print("Starting the vehicle...")

class Car(Vehicle):
    def start(self):
        print("Starting the car engine...")

class Bike(Vehicle):
    def start(self):
        print("Starting the bike engine...")

# Function accepting base class but works for any subclass
def vehicle_start(vehicle: Vehicle):
    vehicle.start()

# Dynamic behavior (method resolved based on object type)
vehicle_start(Car())
vehicle_start(Bike())
vehicle_start(Vehicle())

````

### **Example 3: Real-World Analogy (Payment System)**

````python
class Payment:
    def pay(self):
        print("Processing generic payment...")

class CreditCardPayment(Payment):
    def pay(self):
        print("Processing Credit Card payment...")

class PayPalPayment(Payment):
    def pay(self):
        print("Processing PayPal payment...")

# Function demonstrating dynamic polymorphism
def process_payment(payment: Payment):
    payment.pay()

# Same function behaves differently depending on the object
process_payment(CreditCardPayment())
process_payment(PayPalPayment())
````

## üß≠ Summary

| **Concept** | **Description** |
|--------------|------------------|
| **Dynamic Polymorphism** | Method resolution happens at runtime |
| **Achieved By** | Method overriding |
| **Base Class Reference** | Can point to subclass objects |
| **Result** | Executes the method of the actual object‚Äôs class |
| **Also Known As** | Runtime Polymorphism |

‚úÖ Enables **flexibility**, **extensibility**, and **clean design** in OOP systems.

## Abstract Method and Class, Empty Class, Data Class, Keyword Arguments


## üß© **1. Abstract Method and Abstract Class**

### **Definition**
- An **Abstract Class** is a class that **cannot be instantiated directly**.
- It is used to **define a common interface** for all subclasses.
- An **Abstract Method** is a method that is declared but has **no implementation** ‚Äî subclasses must override it.
- Achieved using the **`abc` (Abstract Base Class)** module.

### **Why Use Abstract Classes?**
- To **enforce structure** across subclasses.
- Ensures **certain methods are always implemented** in derived classes.
- Promotes **consistency** and **polymorphism** in OOP design.



````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Must be overridden by subclasses

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

# Uncommenting below line will raise an error (cannot instantiate abstract class)
# s = Shape()

c = Circle(5)
s = Square(4)
print("Circle Area:", c.area())
print("Square Area:", s.area())
````

## üß± **2. Empty Class**

### **Definition**
- A class with **no properties or methods**.
- Created using the **`pass`** keyword.
- Acts as a **placeholder**, or can be used for **dynamic attribute assignment** later.

````python
class Empty:
    pass

obj = Empty()
obj.name = "Umer"
obj.age = 25
print(f"Name: {obj.name}, Age: {obj.age}")
````

## üßÆ **3. Data Class**

### **Definition**
- Introduced in **Python 3.7**, the `@dataclass` decorator simplifies class creation for storing data.
- Automatically generates:
  - `__init__()`
  - `__repr__()`
  - `__eq__()`, etc.
- Great for classes that mainly hold **attributes without complex behavior**.

````python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int
    grade: str

# Automatically has __init__, __repr__, and __eq__
s1 = Student("Ali", 20, "A")
s2 = Student("Ali", 20, "A")

print(s1)            # Prints readable object
print(s1 == s2)      # True (auto-generated equality check)
````

## ‚öôÔ∏è **4. Keyword Arguments**

### **Definition**
- Allow passing arguments to a function using **parameter names**, not just order.
- Improves **readability** and **clarity** of code.
- Can use:
  - **Default arguments**
  - **`*args` (variable positional arguments)**
  - **`**kwargs` (variable keyword arguments)**

````python
# Example 1: Regular keyword arguments
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce(age=25, name="Umer")

# Example 2: Using *args and **kwargs
def show_details(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

show_details("AI Developer", "Pakistan", company="RunTribe", role="Engineer")
````

##  Measure of Central Tendency (Mean, Median, Mode)



## üß© 1Ô∏è‚É£ What is Central Tendency?

### **Definition**
- **Central Tendency** is a statistical concept used to identify a **single value** that represents the **center point or typical value** of a dataset.
- It helps summarize a large dataset with one representative value.



## üßÆ 2Ô∏è‚É£ Main Measures of Central Tendency

| **Measure** | **Definition** | **Formula / Method** | **Use Case** |
|--------------|----------------|----------------------|---------------|
| **Mean** | The **average** of all data points. | Sum of values √∑ Number of values | When data is evenly distributed |
| **Median** | The **middle value** when data is ordered. | Sort data ‚Üí Select middle value | When data has outliers or skewed distribution |
| **Mode** | The **most frequently occurring value**. | Value with highest frequency | When identifying common category/value |



## üß† 3Ô∏è‚É£ When to Use Which Measure

| **Situation** | **Best Measure** | **Reason** |
|----------------|------------------|-------------|
| Data is **normally distributed** | Mean | Represents overall balance point |
| Data has **outliers or skewed values** | Median | Not affected by extreme values |
| Data is **categorical or repetitive** | Mode | Shows most common occurrence |



## üßÆ 4Ô∏è‚É£ Example

For the dataset:  
> [2, 3, 5, 7, 8, 10, 10, 12]

- **Mean** = (2 + 3 + 5 + 7 + 8 + 10 + 10 + 12) / 8 = 7.125  
- **Median** = (7 + 8) / 2 = 7.5  
- **Mode** = 10  



## üß≠ Summary

| **Concept** | **Represents** | **Affected by Outliers?** |
|--------------|----------------|----------------------------|
| **Mean** | Arithmetic average | ‚úÖ Yes |
| **Median** | Middle value | ‚ùå No |
| **Mode** | Most frequent value | ‚ùå No |

‚úÖ **Together, these three measures give a complete picture** of how your data values are distributed around the center.


````python
import numpy as np
import pandas as pd
from scipy import stats

# Example dataset
data = [2, 3, 5, 7, 8, 10, 10, 12]

# Calculate Mean, Median, and Mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = stats.mode(data, keepdims=True).mode[0]

print(f"Data: {data}")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")

# Create a small dataframe for display
df = pd.DataFrame({'Data': data})
df.plot(kind='hist', bins=8, rwidth=0.8, color='lightblue', edgecolor='black', title='Distribution of Data')

# Mark central tendency values
import matplotlib.pyplot as plt
plt.axvline(mean_val, color='red', linestyle='--', label=f"Mean: {mean_val:.2f}")
plt.axvline(median_val, color='green', linestyle='--', label=f"Median: {median_val:.2f}")
plt.axvline(mode_val, color='orange', linestyle='--', label=f"Mode: {mode_val}")
plt.legend()
plt.show()
````

#   **Day 5 : Overview**

##  Measure of Dispersion (Variance, Standard Deviation, Coefficient of Variation, Skewness, and Kurtosis)



## üß© 1Ô∏è‚É£ What is Dispersion?

### **Definition**
- **Dispersion** refers to the **spread or variability** of data points around a measure of central tendency (like the mean).  
- It tells us **how consistent or scattered** the data is.



## üßÆ 2Ô∏è‚É£ Key Measures of Dispersion

| **Measure** | **Definition** | **Indicates** |
|--------------|----------------|----------------|
| **Variance (œÉ¬≤)** | Average of squared deviations from the mean | Overall spread of data |
| **Standard Deviation (œÉ)** | Square root of variance | How much data deviates from the mean |
| **Coefficient of Variation (CV)** | Ratio of SD to Mean √ó 100 | Relative variability (in %) |
| **Skewness** | Degree of asymmetry in data distribution | Direction of skew (left/right) |
| **Kurtosis** | Degree of peakedness or flatness of distribution | Shape of distribution tails |



## üß† 3Ô∏è‚É£ Formula Overview

| **Measure** | **Formula** |
|--------------|-------------|
| **Variance** | œÉ¬≤ = Œ£(x - Œº)¬≤ / n |
| **Standard Deviation** | œÉ = ‚àöVariance |
| **Coefficient of Variation** | CV = (œÉ / Œº) √ó 100 |
| **Skewness** | \( \text{Skewness} = \frac{\sum (x - \mu)^3 / n}{\sigma^3} \) |
| **Kurtosis** | \( \text{Kurtosis} = \frac{\sum (x - \mu)^4 / n}{\sigma^4} \) |



## üìä 4Ô∏è‚É£ Interpretation

### **A. Variance / Standard Deviation**
- **Low value:** Data points are close to mean (consistent data)
- **High value:** Data points are widely spread (inconsistent data)

### **B. Coefficient of Variation (CV)**
- Expresses **relative variability** ‚Äî useful for comparing datasets with different units or scales.  
> Smaller CV = More consistent data.

### **C. Skewness**
| Type | Description | Shape |
|------|--------------|--------|
| **Positive Skew (Right)** | Tail on right side | Mean > Median |
| **Negative Skew (Left)** | Tail on left side | Mean < Median |
| **Zero Skew** | Symmetrical distribution | Mean = Median |

### **D. Kurtosis**
| Type | Description | Shape |
|------|--------------|--------|
| **Leptokurtic** | Sharp peak, heavy tails | > 3 |
| **Mesokurtic** | Normal peak | = 3 |
| **Platykurtic** | Flat peak, light tails | < 3 |



## üß≠ Summary

| **Measure** | **Meaning** | **Ideal Value** |
|--------------|--------------|----------------|
| **Variance / SD** | Data spread | Low preferred |
| **CV** | Relative variation | < 20% (stable data) |
| **Skewness** | Asymmetry | ‚âà 0 (symmetrical) |
| **Kurtosis** | Peak shape | ‚âà 3 (normal) |

‚úÖ **Together, these measures explain both the spread and shape of your dataset‚Äôs distribution.**


````python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# Example dataset
data = [12, 15, 18, 19, 21, 22, 25, 28, 30, 35, 40]

# Convert to Pandas Series
series = pd.Series(data)

# Calculations
mean = np.mean(series)
variance = np.var(series, ddof=0)         # Population variance
std_dev = np.std(series, ddof=0)          # Population standard deviation
cv = (std_dev / mean) * 100               # Coefficient of variation
skewness = skew(series)
kurt = kurtosis(series, fisher=False)     # Pearson‚Äôs definition (Normal = 3)

# Display results
print("Dataset:", data)
print(f"Mean: {mean:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Coefficient of Variation (CV): {cv:.2f}%")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurt:.2f}")

# Plot distribution
plt.figure(figsize=(8, 4))
plt.hist(series, bins=8, color='lightblue', edgecolor='black', rwidth=0.8)
plt.title("Data Distribution")
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.legend()
plt.show()
````

## üìò Z-Score, Percentile, and Quartile

## 1Ô∏è‚É£ Z-Score (Standard Score)
The **Z-Score** indicates how many **standard deviations** a data point is from the **mean** of the dataset.

### **Formula**
\[
Z = \frac{X - \mu}{\sigma}
\]

Where:  
- \( X \) = data point  
- \( \mu \) = mean of dataset  
- \( \sigma \) = standard deviation  

### **Interpretation**
- **Z = 0** ‚Üí data point is exactly at the mean.  
- **Z = +1** ‚Üí one standard deviation above the mean.  
- **Z = -1** ‚Üí one standard deviation below the mean.  

‚úÖ **Use:** Z-scores are commonly used for **standardization** and **outlier detection**.



## 2Ô∏è‚É£ Percentiles
A **percentile** indicates the **position** of a value in a dataset relative to others.  
For example, if a student is in the **90th percentile**, it means they scored **better than 90%** of all students.

### **Formula**
\[
P_k = \frac{k}{100} \times (N + 1)
\]
Where:
- \( P_k \) = k-th percentile  
- \( N \) = number of observations  

‚úÖ **Use:** Helpful in **ranking**, **comparisons**, and **statistical summaries** (e.g., SAT scores).



## 3Ô∏è‚É£ Quartiles
**Quartiles** divide the data into **four equal parts** (each 25%).  

| Quartile | Description | Position |
|-----------|--------------|-----------|
| **Q1** | Lower quartile (25th percentile) | Median of the lower half |
| **Q2** | Median (50th percentile) | Median of dataset |
| **Q3** | Upper quartile (75th percentile) | Median of upper half |

### **Interquartile Range (IQR)**
\[
IQR = Q3 - Q1
\]
‚úÖ **Use:** Measures the **spread** of the middle 50% of data and is robust to outliers.



## üí° Summary
| Concept | Meaning | Main Use |
|----------|----------|-----------|
| **Z-Score** | Relative position using mean & SD | Standardization |
| **Percentile** | Relative ranking in % | Comparison |
| **Quartile** | Divides data into 4 parts | Spread measurement |



## üß† Pro Tip
In Python, you can compute:
- Z-Score ‚Üí `scipy.stats.zscore(data)`  
- Percentiles ‚Üí `np.percentile(data, 90)`  
- Quartiles ‚Üí `np.percentile(data, [25, 50, 75])`


````python
import numpy as np
from scipy import stats

# Sample data
data = np.array([12, 15, 20, 25, 30, 35, 40])

# Z-Score
z_scores = stats.zscore(data)

# Percentiles
percentile_90 = np.percentile(data, 90)

# Quartiles
Q1, Q2, Q3 = np.percentile(data, [25, 50, 75])
IQR = Q3 - Q1

print("Z-Scores:", np.round(z_scores, 2))
print("90th Percentile:", percentile_90)
print("Quartiles -> Q1:", Q1, ", Q2 (Median):", Q2, ", Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
````
