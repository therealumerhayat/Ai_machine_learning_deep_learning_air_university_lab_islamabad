##  1. Salary Distribution by Rank and Discipline (Box Plot)

**Plot Type:** Box plot

**Question:** How does salary vary by both rank and discipline?

**Visualization:** Create a box plot with salary on the y-axis, and separate the data by rank and discipline on the x-axis.

```python
#importing liabraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#loading data
df = pd.read_csv("D:\Ai_machine_learning_deep_learning_air_university_lab_islamabad\data\Salaries.csv")
df
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='rank', y='salary', hue='discipline', data=df)
plt.title('Salary Distribution by Rank and Discipline')
plt.xlabel('Rank')
plt.ylabel('Salary')
plt.legend(title='Discipline')
plt.show()
```

##  2. Average Salary by Rank and Gender (Bar Plot)

**Plot Type:** Grouped bar plot

**Question:** How does the average salary differ by academic rank and gender?

**Visualization:** Create a bar plot with rank on the x-axis, average salary on the y-axis, and use different bars for male and female.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='rank', y='salary', hue='sex', data=df)
plt.title('Average Salary by Rank and Gender')
plt.xlabel('Rank')
plt.ylabel('Salary')
plt.legend(title='Gender')
plt.show()
```

##  3. Salary vs. Years of Service (Scatter Plot with Regression Line)

**Plot Type:** Scatter plot with regression line

**Question:** What is the relationship between years of service and salary?

**Visualization:** Plot service on the x-axis and salary on the y-axis, with a regression line to see the trend.

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['service'], df['salary'], alpha=0.5)

m, b = np.polyfit(df['service'], df['salary'], 1)
plt.plot(df['service'], m*df['service'] + b, color='red', linewidth=2)


plt.title('Salary vs Years of Service')
plt.xlabel('Years of Service')
plt.ylabel('Salary')
plt.show()
```

```python
#optional
sns.regplot(x='service', y='salary', data=df, line_kws={'color': 'red'})
```

##  4. Distribution of Years Since PhD by Rank (Violin Plot)

**Plot Type:** Violin plot

**Question:** What is the distribution of years since PhD across different academic ranks?

**Visualization:** Use a violin plot with phd on the y-axis and rank on the x-axis to show the distribution.

```python
plt.figure(figsize=(12, 6))

sns.violinplot(x='rank', y='phd', data=df)
plt.title('Distribution of Years Since PhD by Rank')
plt.xlabel('Rank')
plt.ylabel('Years Since PhD')
plt.show()
```

##  5. Gender Distribution within Each Discipline (Stacked Bar Plot)

**Plot Type:** Stacked bar plot

**Question:** What is the gender distribution within each discipline?

**Visualization:** Create a stacked bar plot with discipline on the x-axis and counts on the y-axis, stacking by sex.

```python
pivot_df = df.pivot_table(index='discipline', columns='sex', values='salary', aggfunc='count')
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Gender Distribution within Each Discipline')
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.show()
```

##  6. Salary by Years Since PhD and Gender (Line Plot)

**Plot Type:** Line plot with multiple lines

**Question:** How does salary differ for individuals with the same rank but different years since PhD?

**Visualization:** Plot phd on the x-axis and salary on the y-axis, with separate lines for male and female.

```python
plt.figure(figsize=(20, 6))
sns.lineplot(data=df, x='phd', y='salary', hue='sex', marker='o')
plt.xticks(range(0, df['phd'].max()+1, 5))
plt.title('Salary by Years Since PhD and Gender')
plt.xlabel('Years Since PhD')
plt.ylabel('Salary')
plt.show()
```

##  7. Correlation Matrix (Heatmap)

**Plot Type:** Heatmap

**Question:** What are the correlations between salary, years since PhD, and years of service?

**Visualization:** Create a heatmap showing the correlation coefficients between salary, phd, and service.

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df[["salary", "phd", "service"]].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

##  9. Salary by Discipline and Years of Service (Facet Grid with Scatter Plots)

**Plot Type:** Facet grid of scatter plots

**Question:** What is the relationship between years of service and salary for different disciplines?

**Visualization:** Create scatter plots for service vs. salary, with separate plots for each discipline.

```python
g = sns.FacetGrid(df, col='discipline', height=6, aspect=1.2)
g.map_dataframe(sns.scatterplot, x='service', y='salary')
g.set_axis_labels('Years of Service', 'Salary')
g.fig.suptitle('Salary by Discipline and Years of Service', y=1.03)
plt.show()
```

##  10. Years Since PhD and Service (Joint Distribution)

**Plot Type:** Joint distribution plot

**Question:** What is the relationship between years since PhD and years of service?

**Visualization:** Plot `phd` on the x-axis and `service` on the y-axis, showing both the scatter plot and histograms for each axis.

```python
sns.jointplot(data=df, x='phd', y='service', kind='scatter', height=7,ratio=12, marginal_kws=dict(bins=20, fill=True))
plt.suptitle('Relationship Between Years Since PhD and Years of Service', y=1.02)
plt.show()
```

##  12. Average Salary by Years of Service (Bar Plot)

**Plot Type:** Bar plot

**Question:** How does the average salary change with years of service?

**Visualization:** Group the data by service and plot the average salary for each group.

```python
df_group = df.groupby(['service'])['salary'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=df_group, x='service', y='salary', color='skyblue')

plt.title('Average Salary by Years of Service')
plt.xlabel('Years of Service')
plt.ylabel('Average Salary')
plt.show()
```

##  14. Proportion of Gender in Each Rank (Pie Chart)

**Plot Type:** Pie chart

**Question:** What is the proportion of males and females within each rank?

**Visualization:** Create a pie chart for each rank to show the proportion of sex.

```python
ranks = df['rank'].unique()
plt.figure(figsize=(12,6))

for i, r in enumerate(ranks, 1):
    plt.subplot(1, len(ranks), i)
    gender_count = df[df['rank'] == r]['sex'].value_counts()
    plt.pie(gender_count, labels=gender_count.index, autopct="%1.1f%%", startangle=90)
    plt.title(r)

plt.suptitle('Proportion of Gender in Each Rank')
plt.show()
```
