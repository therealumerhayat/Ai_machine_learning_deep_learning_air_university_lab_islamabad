```python
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['Cause Name']])

encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(["Cause Name"]))
df = pd.concat([df, encoded_df], axis=1).drop("Cause Name", axis=1)
```
